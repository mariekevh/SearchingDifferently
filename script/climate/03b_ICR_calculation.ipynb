### 1. Calculation of ICR

```python
def ICB_calculation(df1, df2, varname, conditional=None):
    tmpdf = pd.concat([df1[varname], df2[varname]], axis=1)
    if conditional:
        print('Conditional on...', conditional)
        index_conditional=pd.concat([df1[conditional], df2[conditional]], axis=1, keys=["MVH", "DT"]).query("MVH==1 & DT==1").index.to_list()
        tmpdf = tmpdf[tmpdf.index.isin(index_conditional)]
    matrix = np.array([tmpdf.iloc[:,0], tmpdf.iloc[:,1]])
    alpha = krippendorff.alpha(reliability_data=matrix, level_of_measurement='nominal')
    print(f'The intercoderreliability for {varname} is {alpha}')
    alpha = round(alpha, 3)
    return {varname:alpha}
```

```python
# VALID
icr1 = ICB_calculation(sample1, sample2, 'GELDIG')
```

```python
varnames1 = ['SPECIFIEK', 'EFFECTIEF', 'VRAAG', 'TAAL']
```

```python
icr2={}
for varname in varnames1:
    icr2.update(ICB_calculation(sample1, sample2, varname, conditional='GELDIG'))
```

```python
varnames2=['INDIVIDUEEL_GEDRAG', 'VERVOER', 'INDUSTRIE', 'VOEDSELINDUSTRIE', 'STIKSTOF', 'CO2',
           'ENERGIE', 'INFO_OPLOSSINGEN', 'INFO_STATUS_QUO', 'OORZAKEN', 'SCEPTICI', 
           'SCEPTICI_PERSONEN', 'GEVOLG', 'POLITIEK_BELEID', 'MILIEU', 'DISCUSSIE', 'ACTIVISME']
           #'ONGECATEGORISEERD']
```

```python
icr3={}
for varname in varnames2:
    icr3.update(ICB_calculation(sample1, sample2, varname, conditional='GELDIG'))
```

```python
result = pd.DataFrame({**icr1, **icr2, **icr3}, index=["Krippendorff Alpha"]).T
result.to_latex(PATH+"/report/tables/climate/IRC_table.tex")
result
```

### 2. Inspect disagreement

```python
sample1.columns = [c+"_MVH" for c in sample1.columns]
sample2.columns = [c+"_TDV" for c in sample2.columns]
```

```python
samples = pd.concat([sample1, sample2],axis=1)
samples = samples.reindex(sorted(samples.columns), axis=1)
```

```python
#samples
```

```python
varstotal=["GELDIG"]+varnames1+varnames2
```

```python
disagreement = {}
```

```python
filterdisagreements = samples.query('GELDIG_MVH!=GELDIG_TDV').index.to_list()
disagree = samples.loc[filterdisagreements,]
disagree = disagree[['GELDIG_MVH', 'GELDIG_TDV', 'Comments_TDV', 'Comments_MVH']]
disagreement.update({'GELDIG':disagree})
```

```python
disagree
```

```python
for varname in varnames1:
    cols = [c for c in samples.columns if c.startswith(varname)]
    index_conditional=samples.query("GELDIG_MVH==1 & GELDIG_TDV==1").index.to_list()
    df = samples[samples.index.isin(index_conditional)]
    filterdisagreements = df.query(f'{cols[0]}!={cols[1]}').index.to_list()
    disagree = df.loc[filterdisagreements,]
    disagree = disagree[cols+['Comments_TDV', 'Comments_MVH']]
    disagreement.update({varname:disagree})
```

```python
for varname in varnames2:
    cols = [c for c in samples.columns if c.startswith(varname)]
    index_conditional=samples.query("SPECIFIEK_MVH==1 & SPECIFIEK_TDV==1").index.to_list()
    df = samples[samples.index.isin(index_conditional)]
    filterdisagreements = df.query(f'{cols[0]}!={cols[1]}').index.to_list()
    disagree = df.loc[filterdisagreements,]
    disagree = disagree[cols+['Comments_TDV', 'Comments_MVH']]
    disagreement.update({varname:disagree})
```
